```python
df["Booking Charge per km"]=round(df['Booking Value']/df['Ride Distance'],2)
```

```python
df["Date"]=pd.to_datetime(df["Date"], errors='coerce')
```

```python
df['Time'] = pd.to_datetime(df['Time'], format='%H:%M:%S', errors='coerce').dt.time
```

```python
df["Date Time"]=pd.to_datetime(df["Date"].astype(str) + " " + df["Time"].astype(str), errors='coerce')
```

```python
df.head(2)
```

```python
df["hours"]=df["Date Time"].dt.hour
```

```python
df["Day"]=df["Date Time"].dt.day
```

```python
df["Month"]=df["Date Time"].dt.month
```

```python
df["WeekDay"]=df["Date Time"].dt.dayofweek
```

```python
df[["Date Time","hours","Day","Month","WeekDay"]].head()
```

```python
df.isnull().sum()
```

```python
df["Avg VTAT"]=df["Avg VTAT"].fillna(df["Avg VTAT"].median())
```

```python
df["Avg CTAT"]=df["Avg CTAT"].fillna(df["Avg CTAT"].median())
```

```python
df["Booking Value"]=df['Booking Value'].fillna(0)
```

```python
df[["Ride Distance","Incomplete Rides"]]=df[["Ride Distance","Incomplete Rides"]].fillna(0)
```

```python
df.head(2)
```

```python
import numpy as np
```

```python
df[["Customer Rating","Driver Ratings"]]=np.round(df[["Customer Rating","Driver Ratings"]].fillna(df[["Customer Rating","Driver Ratings"]].mean()),1)
```

```python
df["Booking Charge per km"]=df["Booking Charge per km"].fillna(0)
```

```python
df.dtypes
```

```python
df[["Payment Method","Reason for cancelling by Customer","Driver Cancellation Reason","Incomplete Rides Reason"]]=df[["Payment Method","Reason for cancelling by Customer","Driver Cancellation Reason","Incomplete Rides Reason"]].fillna("Not Known")
```

```python
df[["Cancelled Rides by Driver","Cancelled Rides by Customer"]]=df[["Cancelled Rides by Driver","Cancelled Rides by Customer"]].fillna(0)
```

```python
df.isnull().sum()
```

```python
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
sns.countplot(data=df,x="Month" , order=sorted(df["Month"].unique()))
plt.title("Count of Rides per month")
plt.show()
plt.figure(figsize=(2,3))
```

```python
ax=sns.countplot(data=df, x="Reason for cancelling by Customer",order=df["Reason for cancelling by Customer"].value_counts().index,color='violet')
plt.title("Reasons for cancelled rides by customer")
plt.xticks(rotation=70)
plt.figure(figsize=(7,4))
for container in ax.containers:
    ax.bar_label(container)
plt.show()
 


```

```python
df['target_customer_cancelled'] = df['Cancelled Rides by Customer'].astype(int)
```

```python
df['target_customer_cancelled'].value_counts(normalize=True)
```

```python
sns.barplot(data=df, x='Vehicle Type', y='target_customer_cancelled')
plt.title("Customer Cancellation Rate by Vehicle Type")
plt.xticks(rotation=35)
plt.show()
```

```python
sns.boxplot(data=df, x='target_customer_cancelled', y='Driver Ratings')
plt.title("Driver Ratings vs Customer Cancellation")
plt.xticks([0, 1], ['Not Cancelled', 'Cancelled'])
plt.show()
```

```python
heatmap_data = df.pivot_table(
    index='hours',
    columns='WeekDay',
    values='target_customer_cancelled',
    aggfunc='mean'
)
sns.heatmap(heatmap_data, cmap="YlGnBu", annot=True, fmt=".2f")
plt.title("Cancellation Rate in Hours and Day of a week")
plt.ylabel("hours")
plt.xlabel("Weekday (0=Mon, 6=Sun)")
plt.show()
```

```python
top_pickups = df['Pickup Location'].value_counts().nlargest(10).index
df_top_pickups = df[df['Pickup Location'].isin(top_pickups)]
sns.barplot(data=df_top_pickups, x='Pickup Location', y='target_customer_cancelled')
plt.title("Cancellation Rate by Top 10 Pickup Locations")
plt.xticks(rotation=45)
plt.figure(figsize=(2,3))
plt.show()
```

```python
sns.barplot(data=df, x='WeekDay', y='target_customer_cancelled')
plt.title("Customer Cancellation Rate of Week days")
plt.xticks(ticks=range(7), labels=["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"])
plt.show()
```
